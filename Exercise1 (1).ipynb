## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
import numpy as np

# sample data
x = np.linspace(0, 5, 11)
y = x

# create figure
fig = plt.figure()

# add axes: [left, bottom, width, height]
ax = fig.add_axes([0, 0, 1, 1])

# plot on the axes
ax.plot(x, y)

# labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')

plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.arange(0, 101)
y = x * 2
z = x ** 2

# Create figure
fig = plt.figure(figsize=(8,4))

# Left plot
ax1 = fig.add_axes([0.0, 0.0, 0.45, 1.0])
ax1.plot(x, y, 'b--')   # blue dashed
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

# Right plot
ax2 = fig.add_axes([0.55, 0.0, 0.45, 1.0])
ax2.plot(x, z, 'r')     # red line
ax2.set_xlabel("X")
ax2.set_ylabel("Z")

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import numpy as np
import matplotlib.pyplot as plt

# data
x = np.arange(0, 101)
y = x * 2
z = x ** 2

# -> only change: use 1 row, 2 cols (side-by-side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# left: blue solid, a bit thicker
ax1.plot(x, y, color='blue', linestyle='-', linewidth=2)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# right: red dashed, thicker
ax2.plot(x, z, color='red', linestyle='--', linewidth=3)
ax2.set_xlabel('X')
ax2.set_ylabel('Z')

plt.tight_layout()
plt.show()
```
